```python
import os
cnt =0
DATA=[]
datas = open('appbp','r').read()
data=datas.replace('@','---INDEX---\n@')
routes = data.split('---INDEX---')
for route in routes:
    if '@' in route:DATA.append([route])
    lines = route.split("'")
    for line in lines:
        if '.html' in line:
            cnt=cnt+1
            html_path = os.path.join(os.getcwd(), 'templates/', line)
            if len(html_path)>8 and len(html_path)<100:
                print(f'HTML: {html_path}')
                htm = open(f'{html_path}','r').read()
                DATA.append([htm])
                
all_ = open("DATA.txt","w")
for line in DATA:
    cnt = cnt +1
    line = str(line).replace('\\n','\n')
    all_.write(line)                
```
